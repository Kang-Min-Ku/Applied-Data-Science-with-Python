## Regular Expression

Very powerful when finding pattern.

[Regex Documentation](https://docs.python.org/3/library/re.html)

```python
text12 = '"Ethics are built right into the ideals and objectives of the United Nations" #UNSG @ NY Society For Ethical Culture bit.ly/2guVelr @UN @UN_Women'
[word for word in text12.split(' ') if word.startswith('@')]
```

```python
import re
[word for word in text12.split(' ') if re.search('@\w+',word)]
```

```python
text13 = 'ouagadougou'
re.findall(r'[aeiou]',text13)
```

'r' in front of string is symbol of **raw string**.

It doesn't interpret back slash, so you can represent \\d as \d.

```python
dateStr = '23 Oct 2002'
re.findall(r'\d{2} (Jun|Fec|Mar|Oct)',dateStr) #['Oct']
re.findall(r'\d{2} (?:Jun|Fec|Mar|Oct)',dateStr) #['23 Oct']
```

Well, the difference here is that this bracket sign has two meanings. In regular expressions, when you use the bracket, it also indicates scoping. S**o it says, I want to pull out only something that matched that part. It doesn't match the whole string**, but it pulls out and gives you back only the thing that matched between Jan, Feb, March up to December. That's what gave me October as O-C-T.

### Internationalization and Issues with Non ASCII Characters

**ASCII** : American Standard Code for Information Interchange
 - 7-bit character encoding
 - 0x00 - 0x7F
 - Work relatively well for English typewriting, but don't cover everything
 - Only for english
 
But, there are many languages, many emoticons and many shapes(?) in the world.
2^7 is not enough

**[Unicode](https://ko.wikipedia.org/wiki/%EC%9C%A0%EB%8B%88%EC%BD%94%EB%93%9C)** : Industry standard for encoding and representing text

 - Over 128,000 characters from 130+ scripts and symbol sets
 - Can be implemented by different character endings
     * UTF-8 : Ony byte to up to four bytes (most common)
     * UTF-16 : One or two 16-bit code units

**[UTF-8](https://ko.wikipedia.org/wiki/UTF-8)**
 - Unicode Transformational Format -8 bits
 - Variable length encoding: One to four bytes
 - Backward compatible with ASCII
   - One byte codes same as ASCII
 - Dominant character encoding for the Web
 - In python
   - Python 3 : default
   - Python 2 : Need -*- coding:utf-8-*-
